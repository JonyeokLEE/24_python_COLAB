'''
[최소제곱법 사용]
최소제곱법은 클러스터를 어떤 직선으로 근사했을 때 발생하는 오차의 제곱합을 최소화한다.

E = (sigma i=1 to n)(a*xi - yi)^2
여기서 E를 최소화 한다는 것.
E를 a에 대한 함수로 생각하고 E를 a에 관해서 미분한 함수가 0이 되는 조건을 찾는다.

a = (x^T*y) /(||x||)^2
'''

```python
import numpy as np
import matplotlib.pyplot as plt

def reg1dim1(x,y):
  a = np.dot(x,y)/(x**2).sum()
  return a

x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a = reg1dim1(x,y)

plt.scatter(x,y,color="k")
xmax = x.max()
plt.plot([0,xmax],[0,a*xmax], color = "r")
plt.show()
```
