```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3])
y = np.array([3,7,4,8])

plt.plot(x,y,color="r")
plt.show()
```

```python
plt.scatter(x,y,color="r")
plt.show()
```
