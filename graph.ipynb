```python
import numpy as np
from matplotlib import pyplot as plt

epoch = np.arange(1,100)
train_loss = 1.5/(epoch)
val_loss = .3+1/(epoch)
#loss가 작으면 실제값과 예측값의 차이가 작다 -> 학습이 잘 되었다
```

```python
plt.figure(figsize=(10,5)) #figure는 그래프의 틀을 담당하며, figsize(너비,높이)로 사이즈 조절
plt.plot(train_loss,'r-*')
plt.plot(val_loss,'b-*') #해당 변수의 값을 r-(red)/b-(blue) 값은 *로
plt.legend(['train','validation']) #차트의 범례
plt.title("MNIST") #타이틀
plt.xlabel("epoch")
plt.ylabel("loss") #x축과 y축
plt.show() #그래프 출력
```
